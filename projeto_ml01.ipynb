## Colunas do Dataset

**age**: Idade.

**workclass**: Classe trabalhadora. Se a pessoa é funcionário público, autônomo, empresário etc.

**education**: Níveis de Escolaridade.

**education-num**: Código do Nível de Escolaridade.

**marital-status**: Estado Civil.

**occupation**: Ocupação (Suporte técnico, Conserto de arte, Outros serviços, Vendas, Exec-managerial, Prof-specialty, Manipuladores de limpeza, Machine-op-inspct, Adm-clerical, Pesca agropecuária, Transport-moving, Priv-house-serv, Serviço de Proteção, Forças Armadas)

**relationship**: Relacionamento (Esposa, Filha Própria, Marido, Não-familiar, Outro-parente, Solteira.)

**race**: Raça (Preto, branco, asiático-Pac-Islander, Amer-Indian-Eskimo, Outro.)

**sex**: Sexo(Feminino, Masculino).

**capital-gain**: Ganho de Capital.

**capital-loss**: Perda de Capital.

**hours-per-week**: Horas de trabalho por Semana.

**native-country**: País Nativo (Estados Unidos, Camboja, Inglaterra, Porto Rico, Canadá, Alemanha, EUA (Guam-USVI-etc), Índia, México, Portugal etc)

####  <span style="color:Blue">  Classificação e Avaliação de Modelos de Machine Learning </span>

Foram usados os algoritmos supervisionados para classificação o KNN e SVC.



```python
# Algoritmo de Machine Learning KNN
# Separa os dados e a classe
# Cria o dataframe X com todos os dados removendo a classe
X = df.drop('classe',axis=1)
```

```python
Y = df['classe']
```

```python
# split into training e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
# Aplicando o Machine Learning
# treinando o modelo KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Cria o objeto KNN
knn = KNeighborsClassifier()
```

```python
# treina os dados 
knn.fit(X_train, y_train)
```

```python
# atribui as classificações a variavel resultados
resultados = knn.predict(X_test)
```

```python
# carrega o pacote accuracy_score ( vai verificar a acuracia dos resultados)
from sklearn.metrics import accuracy_score
```

```python
# cria a variável score
score = accuracy_score(y_test,resultados)
```

```python
# imprime score (acurácia)
score
```

```python
# Tecnicas de Validação
# Matriz de Confusão
print(pd.crosstab(y_test,resultados,rownames=['Real'],colnames=['Predito'],margins=True))
```

```python
# Metricas de Classificação
from sklearn import metrics
print(metrics.classification_report(y_test,resultados))
```

```python
# Algoritmo de Machine Learning SVM
from sklearn.model_selection import  cross_val_predict
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import  svm
```

```python
# separa os dados de treino
treino = df.drop('classe',axis=1)
```

```python
# separa os dados da classe
classes = df['classe']
```

```python
# treinando o algoritmo SVM
clf = svm.SVC().fit(treino,classes)
```

```python
clf
```

```python
# atribui as classificações a variavel resultados
resultado_svm = clf.predict(treino)
```

```python
# metricas de classificação 
# cross validation
def Acuracia(clf,X,y):
    resultados = cross_val_predict(clf,X,y,cv=5)
    return metrics.accuracy_score(y,resultados)
```

```python
Acuracia(clf,treino,classes)
```

```python
# Metricas de Avaliação
resultados = cross_val_predict(clf,treino,classes,cv=10)
valor_classes=[0,1]
print (metrics.classification_report(classes,resultados,valor_classes))
```

```python
# Tecnicas de Validação
# Matriz de confusão
print(pd.crosstab(classes,resultado_svm ,rownames=['Real'],colnames=['Predito'],margins=True),' ')
```

```python
# Machine Learning com Pipelines
# importa o pacote do Pipeline
from sklearn.pipeline import  Pipeline
```

```python
# carrega os pacotes
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
from sklearn import svm
```

```python
# Cria os Pipelines
pip_std = Pipeline([('scaler', StandardScaler()),
                    ('KNN', KNeighborsClassifier())
                    ])

pip_minmax = Pipeline([('minmax', MinMaxScaler()),
                    ('KNN', KNeighborsClassifier())
                    ])

pip_std_svm = Pipeline([('scaler', StandardScaler()),
                      ('KNN', svm.SVC())
                       ])

pip_minmax_svm = Pipeline([('minmax', MinMaxScaler()),
                       ('KNN', svm.SVC())
                       ])
```

```python
# Aplicando o MinMaxScaler
pip_minmax.fit(X_train,y_train)
```

```python
# imprime o score
pip_minmax.score(X_test,y_test)
```

```python
# Aplicando StandScaler
# executa o Pipeline pip_std
pip_std.fit(X_train,y_train)
```

```python
# imprime o Score
pip_std.score(X_test,y_test)
```

```python
# aplica SVM com pré-processadores
pip_std_svm.fit(X_train,y_train)
```

```python
# imprime o Score
pip_std_svm.score(X_test,y_test)
```
